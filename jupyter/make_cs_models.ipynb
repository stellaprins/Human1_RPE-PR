## Apply CORDA2 algorithm to create cell-specific models

### load and discretize expression data

```python
import pandas as pd

# set relative path to github repo C:\Users\prins\GitHub\Human1_RPE-PR
import os
os.chdir(r'C:\Users\prins\GitHub\Human1_RPE-PR')

# from RPE_PR data folder, read in expression data
df = pd.read_excel(r'expression data\expression_data.xlsx')
df.head()
```

```python

conditions = df.columns[2:]
df_discretized = df.copy(deep=True)

for c in conditions:
    q25 = df_discretized[c].quantile(0.25)
    q75 = df_discretized[c].quantile(0.75)

    df_discretized.loc[df_discretized[c] < 0.00001, c] = -1  # negative confidence (NC)
    df_discretized.loc[(df_discretized[c] <= q25) & (df_discretized[c] >  0.00001), c] = -10  # 1 'LC'
    df_discretized.loc[(df_discretized[c] <= q75) & (df_discretized[c] > q25), c] = -20  # 2 'MC'
    df_discretized.loc[df_discretized[c] > q75, c] = -30  # 3 'HC'
    

    df_discretized.loc[df_discretized[c] == -10, c] = 1
    df_discretized.loc[df_discretized[c] == -20, c] = 2
    df_discretized.loc[df_discretized[c] == -30, c] = 3

# Replace NA values with 0
df_discretized.fillna(0, inplace=True)

# Display the head of the modified DataFrame
df_discretized.head()
```

### load Human1 template model


```python
from cobra.io import read_sbml_model
mod = read_sbml_model(r'models\template\Human-GEM_28062023.xml')
mod
```

### Map expression to reactions to create cell-specific models


```python
from corda import reaction_confidence
from cobra import Configuration
from corda import CORDA
from cobra.io import write_sbml_model
import datetime
from pathlib import Path

config = Configuration()
config.solver = "glpk"
now = datetime.datetime.now()
date = now.strftime("%Y%m%d")

# create folder if it does not exist Path().cwd() / "models" / "cs_mods" / "date"
folder_path = Path().cwd() / "models" / "cs_mods" / date
folder_path.mkdir(parents=True, exist_ok=True)

i=0
models={} # empty dict to place new models in
for c in conditions:
    mod_c = mod.copy() # make deep copy of basis model 'mod' (Human1) for every cell-specific model
    mod_c.id = c
    gene_conf_dict = dict([(i,x) for i, x in zip(df_discretized.Ensembl, df_discretized[c])]) # gene confidences dictionary
    rxn_conf_dict = {rxn.id:reaction_confidence(rxn, gene_conf_dict) for rxn in mod.reactions} # reaction confidence dictionary
    mod_CORDA = CORDA(mod, rxn_conf_dict) 
    mod_CORDA.build()
    print(c)
    print(mod_CORDA)
    mod_c = mod_CORDA.cobra_model()
    models[i] = mod_c
    model_name = f"{c}.xml"
    write_sbml_model(mod_c, folder_path / model_name)
    i = i + 1
```
