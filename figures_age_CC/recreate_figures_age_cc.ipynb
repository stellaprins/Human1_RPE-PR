````python
# import sns
import seaborn as sns
def dg_axes():
    dg=(.25, .25, .25)
    for ax in plt.gcf().axes:
        ax.spines['bottom'].set_color(dg)
        ax.spines['left'].set_color(dg)
        #set spine linewidth to 1
        ax.spines['bottom'].set_linewidth(1)
        ax.spines['left'].set_linewidth(1)
        for p in ax.patches:
            p.set_edgecolor(dg)
            p.set_linewidth(1)
# set whitegrid style
sns.set_style("whitegrid")
````

## Figure 3




Figure 3. Graph of estimates of choriocapillaris density (ratio) with increasing distance from the foveal centre of the retina at different ages



````python
# FIGURE 3

# print files in folder_Fig3
folder_Fig3 = os.path.join(user_folder, 'Results\Phil\Figures\Figures\Figure 3')
print(os.listdir(folder_Fig3))

# read csv file in folder_Fig3 and convert to dataframe to long format for plotting
csv_file_Fig3 = [f for f in os.listdir(folder_Fig3) if f.endswith('.csv')]
df_Fig3 = pd.read_csv(os.path.join(folder_Fig3, csv_file_Fig3[0]))
df_Fig3_long = pd.melt(df_Fig3, id_vars=['Age', 'Unnamed: 0'])
df_Fig3_long.rename(columns={'Age': 'age'}, inplace=True)
df_Fig3_long.rename(columns={'variable': 'eccentricity (mm)'}, inplace=True)
df_Fig3_long['eccentricity (mm)'] = df_Fig3_long['eccentricity (mm)'].str.replace('X', '').str.replace('mm', '').astype(float)
df_Fig3_long.rename(columns={'value': 'flow deficits (%)'}, inplace=True)

# fix age group names
df_Fig3_long['age'] = df_Fig3_long['age'].astype(str)
old_age = ['25', '35', '45', '55', '65', '75', '85']
new_age = ['19–29', '30–39', '40–49', '50–59', '60–69', '70–79', '80–89']
df_Fig3_long['age'] = df_Fig3_long['age'].replace(old_age, new_age)

# plot Age as hue and eccentricity as x value and value as y value
import seaborn as sns
plt.figure(figsize=(4, 2.5))
sns.lineplot(data=df_Fig3_long, x='eccentricity (mm)', y='flow deficits (%)', hue='age', palette=p , legend=True,marker='o')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1),title='age')
plt.title('flow deficits')
dg_axes()

# convert all flow deficit (%) values to CC density using  CC density = 0.726 – 0.0129 * proportion of flow deficits
df_Fig3_long['CC density (%)'] = (0.726 - 0.0129 * df_Fig3_long['flow deficits (%)']) *100
# plot Age as hue and eccentricity as x value and value as y value
plt.figure(figsize=(4, 2.5))
sns.lineplot(data=df_Fig3_long, x='eccentricity (mm)', y='CC density (%)', hue='age', palette=p , legend=True,marker='o')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1),title='age')
plt.title('CC density')
dg_axes()

# both
plt.figure(figsize=(8, 2))
plt.subplot(1, 2, 1)
sns.lineplot(data=df_Fig3_long, x='eccentricity (mm)', y='flow deficits (%)', hue='age', palette=p , legend=False,marker='o')
plt.title('flow deficits')
plt.subplot(1, 2, 2)
sns.lineplot(data=df_Fig3_long, x='eccentricity (mm)', y='CC density (%)', hue='age', palette=p , legend=True,marker='o')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1.06),title='age')
plt.title('CC density')
plt.ylabel('CC density (%)')
dg_axes()

# save at high resolution in Figures folder folder_to_save_figs
plt.savefig(os.path.join(folder_to_save_figs, 'Figure2.png'), dpi=400, bbox_inches="tight")
````

````python
import pandas as pd
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt


# Rod coverage data in os.path.join(user_folder, 'Results\Phil\Figures\Data')
data_young = pd.read_csv(os.path.join(user_folder, 'Results\Phil\Figures\Data', 'Curcio1993a_F11c_Young.txt'), sep='\t')
data_old = pd.read_csv(os.path.join(user_folder, 'Results\Phil\Figures\Data', 'Curcio1993a_F11c_Old.txt'), sep='\t')

# make dicts with H_rod_cover_young info
H_rod_cover_young ={'ref': 'Curcio1993a', 'xunits': 'mm', 'yunits': '#', 'xlabel': 'eccentricity', 'ylabel': 'rod coverage', 'data': data_young, 'comment': 'mean age = 34'}
H_rod_cover_old = {'ref': 'Curcio1993a', 'xunits': 'mm', 'yunits': '#', 'xlabel': 'eccentricity', 'ylabel': 'rod coverage', 'data': data_old, 'comment': 'mean age = 73'}

# combine data_young and data_old and sort on x
H_rod_cover_temp = pd.concat([data_young, data_old])
H_rod_cover = H_rod_cover_temp.sort_values(by='x')

# plot H_rod_cover
plt.figure(figsize=(4,2.5))
plt.scatter(H_rod_cover['x'], H_rod_cover['y'],alpha=0.5)

# fit polynomial to second degree
x = H_rod_cover['x']
y = H_rod_cover['y']
popt, pcov = np.polyfit(x, y, 2, cov=True)

# plot polynomial
x = np.linspace(0, 8, 100)
y = popt[0]*x**2 + popt[1]*x + popt[2]
plt.plot(x, y, color='black', linewidth=1)

# add text with equation of fit
text = f'y = {popt[0]:.2f}x$^2$ + {popt[1]:.2f}x + {popt[2]:.2f}'
plt.text(0.6, 0.2, text, ha='center', va='center', \
    transform=plt.gca().transAxes, \
    backgroundcolor='white')
plt.title('Rod coverage v eccentricity\nold and young pooled')
plt.xlabel('eccentricity (mm)')
plt.ylabel('rod coverage')
plt.tight_layout()
dg_axes()
````
